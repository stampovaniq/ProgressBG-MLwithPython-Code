```python
import pandas as pd
```

```python
authors = pd.read_json("authors.json")
authors.head()
```

```python
# get unique values for authors and create a new df:
authors_df = pd.DataFrame(authors.author.unique(), columns=["name"])
# authors_df.insert(0, 'id', range(1, 1+len(authors_df)))
authors_df.set_index(["id"])
```

```python
g = authors_df.groupby("name")

for author, author_df in g:
    print(author)
    print(author_df)
```
